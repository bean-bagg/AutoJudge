```python
import pandas as pd
# json to csv
url = "https://raw.githubusercontent.com/AREEG94FAHAD/TaskComplexityEval-24/refs/heads/main/problems_data.jsonl"
df = pd.read_json(url, lines=True)
df.to_csv("Dataset.csv", index=False)
df.head(3)
```

```python
data=pd.read_csv("Dataset.csv")
data["problem_class"].value_counts()
```

```python
data["desc_length"] = data["description"].apply(
    lambda x: len(str(x).split())
)
median_lengths = data.groupby("problem_class")["desc_length"].median()
median_lengths
```

```python
median_scores = data.groupby("problem_class")["problem_score"].median()
median_scores
```

```python
data.fillna("", inplace=True)
data.head()
```

```python
import re
def cleaning(text):
    text=text.lower()
    text=re.sub(r"\s+"," ", text)
    text=text.strip()
    return text
```

```python
data["user_input"] = (
    "__TITLE__" + data["title"].apply(cleaning) + " " +
    "__DESCRIPTION__" + data["description"].apply(cleaning) + " " +
    "__INPUT__" + data["input_description"].apply(cleaning) + " "+
    "__OUTPUT__" + data["output_description"].apply(cleaning))
data["user_input"]
```

```python
import joblib
joblib.dump(data, "training_data.pkl")
```
