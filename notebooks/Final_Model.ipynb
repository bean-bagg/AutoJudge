```python
import joblib
data=joblib.load("training_data.pkl")
tfidf=joblib.load("tfidf.pkl")
data
```

```python
keywords = ["greedy","dp","dynamic programming","tree","graph","dfs","bfs","two pointers","binary search","bitmasks","dfs","combinatorics"]
def extra_features(text):
    return [
        len(text),
        len(text.split()),
        sum(text.count(k) for k in keywords),
        sum(1 for c in text if c in "+-*/%")
    ]
```

```python
import numpy as np
from scipy.sparse import hstack
from sklearn.feature_extraction.text import TfidfVectorizer

x=data["user_input"]
y_class=data["problem_class"]
y_score=data["problem_score"]
```

```python
x_train_tfidf = tfidf.fit_transform(x)
```

```python
x_train_extra = np.array(x.apply(extra_features).tolist())
```

```python
x_train_final = hstack([x_train_tfidf, x_train_extra])
```

```python
# Training Random Forest classifier on full dataset
from sklearn.ensemble import RandomForestClassifier

clf_rf = RandomForestClassifier(
    n_estimators=500,
    max_depth=None,
    class_weight="balanced",
    random_state=42,
    n_jobs=-1
)

clf_rf.fit(x_train_final, y_class)
```

```python
# Training Gradient Boost Regressor on full dataset
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

reg_gbr = GradientBoostingRegressor(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=3,
    random_state=42
)

reg_gbr.fit(x_train_final, y_score)
```

```python
joblib.dump(clf_rf, "final_clf.pkl")
joblib.dump(reg_gbr, "final_reg.pkl")
```
